2. What are the three different types of Boolean operators?

<font color="blue">
    
**Ans:**
    
**AND, OR, and NOT.**
    
The **AND** operator returns True if both operands are **True**, else it returns **False**.
    
The **OR** operator returns True if any of the one operands are **True**, if both operands are it returns **False**.
    
The **NOT** operator returns the opposite Boolean value of its operand. If the operand is **True**, it returns **False**, and if the operand is **False**, it returns **True**.
